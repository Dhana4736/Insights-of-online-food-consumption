### **BAR GRAPH & HISTOGRAM**

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/onlinefoods.csv')

fig, axes = plt.subplots(1, 2, figsize=(18, 6))

# Bar graph of gender distribution
gender_counts = df['Gender'].value_counts()
axes[0].bar(gender_counts.index, gender_counts.values, color=['skyblue', 'lightcoral'])
axes[0].set_title('Gender Distribution')
axes[0].set_xlabel('(Gender)')
axes[0].set_ylabel('(Count)')

# Histogram of age distribution by gender
df[df['Gender'] == 'Male']['Age'].hist(bins=20, color='skyblue', edgecolor='black', alpha=0.5, ax=axes[1], label='Male')
df[df['Gender'] == 'Female']['Age'].hist(bins=20, color='lightcoral', edgecolor='black', alpha=0.5, ax=axes[1], label='Female')
axes[1].set_title('Age Distribution by Gender')
axes[1].set_xlabel('(Age)')
axes[1].set_ylabel('(Frequency)')
axes[1].legend()

plt.tight_layout()
plt.show()
```

### **HEAT MAPS**

```python
import folium
from folium.plugins import HeatMap
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/onlinefoods.csv')

center_location = [df['latitude'].mean(), df['longitude'].mean()]
m = folium.Map(location=center_location, zoom_start=10)

heat_data = [[row['latitude'], row['longitude']] for index, row in df.iterrows()]
HeatMap(heat_data).add_to(m)

m.save('heatmap.html')

m
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/onlinefoods.csv')

plt.figure(figsize=(10, 8))
sns.kdeplot(data=df, x='latitude', y='longitude', cmap='viridis', fill=True, cbar=True)
plt.title('Geospatial Heatmap')
plt.xlabel('Latitude')
plt.ylabel('Longitude')
plt.show()
```

# **CORRELATION ANALYSIS**
